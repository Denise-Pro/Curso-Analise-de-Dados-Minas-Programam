```python
# importamos as bibibliotecas para salvar todas as funções/ferramentas que tem dentro delas na memória 
# fazendo com que fiquem disponíveis para o nosso uso
import requests
from bs4 import BeautifulSoup 
```

```python
# neste trecho a seguir nós faremos o web scrapinc de um arquivo csv (formato colunar separado por vírgulas) que está salvo no meu github

# eu salvo nesta variável 'url' o link http de onde está o meu arquivo
url = "https://github.com/Denise-Pro/Curso-Analise-de-Dados-Minas-Programam/blob/main/def_area_2004_2019.csv"

# o metodo get da biblioteca requests permite que eu faça uma requisição de leitura de um HTML de um site
# e eu estou salvando na variável 'r' o resultado dessa requisição, ou seja, se der certo o resultado dessa leitura estará salvo em r

r = requests.get(url)
```

```python
# r.text me retorna o html que ficou salvo em r
print(r.text)
```

```python
# visto qe a biblioteca BeautifulSoup nos permite extrair os dados de um HTML
# eu passo como parâmetro dentro do () o r.text que corresponde ao HTML que eu quero que seja lido e o método html.parser
# pra
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

Explicarei a Célula a seguir passo a passo na seunda Feira

```python
lista = []
for linha in soup.find_all('table', attrs={"class": "js-csv-data csv-data js-file-line-container"}):
  lista.append(linha)

lista_td = []
lista_areaDesmatadaTotal = []
for linha in soup.find_all('td'):
  lista_td.append(linha)
  for estado in lista_td:
    lista_areaDesmatadaTotal.append(estado.next_element)

len(lista_areaDesmatadaTotal)
```

```python
lista_areaDesmatadaTotal = list(dict.fromkeys(lista_areaDesmatadaTotal))
len(lista_areaDesmatadaTotal)
```
